```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
heart = pd.read_csv("heart_disease_dataset.csv")

heart.head()
```

```python
heart = heart.dropna()
```

```python
heart.isnull().sum()
```

```python
heart = heart.sample(n=80, random_state=42)
```

```python
heart.shape
```

```python
heart.info()
```

```python
sns.countplot(x='target', data=heart)

plt.title("Heart Disease Distribution")
plt.show()
```

```python
heart.columns
```

```python
sns.countplot(x='heart_disease', data=heart)

plt.title("Heart Disease Distribution")
plt.show()
```

```python
X = heart.drop(columns=['heart_disease'])

y = heart['heart_disease']
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
ridge_model = Ridge(alpha=1.0)

ridge_model.fit(X_train, y_train)
```

```python
y_pred = ridge_model.predict(X_test)

y_pred = np.round(y_pred).astype(int)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, cmap="Blues")

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```

```python
heart = pd.read_csv("heart_disease_dataset.csv")
stroke = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
heart = heart.dropna()

heart = heart.sample(n=80, random_state=42)
```

```python
heart['stroke'] = 0
```

```python
stroke = stroke.drop(columns=['id'])
```

```python
stroke['bmi'].fillna(stroke['bmi'].mean(), inplace=True)
```

```python
stroke = pd.get_dummies(stroke, drop_first=True)
```

```python
stroke = stroke.sample(n=80, random_state=42)
```

```python
stroke['heart_disease'] = 0
```

```python
heart_cols = set(heart.columns)
stroke_cols = set(stroke.columns)

missing_in_heart = stroke_cols - heart_cols
missing_in_stroke = heart_cols - stroke_cols
```

```python
for col in missing_in_heart:
    heart[col] = 0

for col in missing_in_stroke:
    stroke[col] = 0
```

```python
stroke = stroke[heart.columns]
```

```python
combined = pd.concat([heart, stroke], ignore_index=True)
```

```python
combined.shape
```

```python
combined['disease'] = 0

combined.loc[
    (combined['heart_disease'] == 1) & (combined['stroke'] == 0),
    'disease'
] = 1

combined.loc[
    (combined['heart_disease'] == 0) & (combined['stroke'] == 1),
    'disease'
] = 2

combined.loc[
    (combined['heart_disease'] == 1) & (combined['stroke'] == 1),
    'disease'
] = 3
```

```python
X = combined.drop(columns=['disease'])

y = combined['disease']
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
ridge_model = Ridge(alpha=1.0)

ridge_model.fit(X_train, y_train)
```

```python
y_pred = ridge_model.predict(X_test)

y_pred = np.round(y_pred).astype(int)
```

```python
y_pred = np.clip(y_pred, 0, 3)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Combined Disease Prediction")

plt.show()
```
